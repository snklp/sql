### **1. SELECT *(columns to view)*  &  FROM *(table they live in).***

**Viewing every column *(using *)***

```python
%%sql

SELECT * FROM inventory   
```

**Viewing specific columns *inventory_id* & *last_update***

```python
%%sql

SELECT inventory_id,
        last_update
    FROM inventory
```

> ***If column name consists of SPACES then write it in ""***

**Renaming columns**

```python
%%sql

SELECT inventory_id AS "Inventory Id",
        last_update AS "Last Update"
    FROM inventory
```

### 2. LIMIT 

**Shows top n rows**

```python
%%sql

SELECT inventory_id AS "Inventory Id",
        last_update AS "Last Update"
    FROM inventory
LIMIT 10
    
```

### 3. WHERE

**Viewing only those rows which satisfy *WHERE* condition**

```python
%%sql

SELECT *
  FROM inventory 
WHERE film_id = 1
```

### 6. Logical Operators
- **LIKE** *(allows to match on similar values rather than exact ones)*
- **IN** *(allows to specify a list of values that to be included in the results)*
- **ILIKE** *(same as LIKE but the matching string isn't case-senstive)*
- **BETWEEN**
- **IS NULL**
- **AND**
- **OR**
- **NOT**

**LIKE**
- *Column name in the WHERE condition (eg- "rating" below) can't be written in single quotes.*

- *Matching String alongwith wildcard(%) (eg- 'PG-%' below) can't be written in double quotes.*

```python
%%sql

SELECT * FROM film WHERE "rating"::text LIKE 'PG-%' LIMIT 5
```

```python
%%sql

SELECT * FROM film WHERE "rating"::varchar LIKE 'PG%' LIMIT 10
```

**IN** *(numerical list of values to be included in the results)*

```python
%%sql

SELECT *
FROM inventory
WHERE film_id in (4,6,3)
```

**IN** *(non-numerical list of values to be included in the results)*

```python
%%sql

SELECT *
FROM film
WHERE rating IN ('NC-17', 'R')
LIMIT 25
```

**BETWEEN** *(used with AND, includes range bounds, eg- 5 & 8 are included below)*

```python
%%sql

SELECT *
FROM inventory
WHERE film_id BETWEEN 5 AND 10
```

***same result as above***

```python
%%sql

SELECT * 
FROM inventory 
WHERE film_id >=5 AND film_id <=10
```

**NULL** *(used along with IS)*

```python
%%sql

SELECT *
FROM address 
WHERE address2 IS NULL
```

**AND** *(used w/ additional AND statements or any other comparison operator as many time we want)*

```python
%%sql

SELECT *
FROM film
WHERE replacement_cost >= 10 AND replacement_cost <= 14
  AND  rating::varchar LIKE 'NC%'
LIMIT 10
```

**OR**

```python
%%sql

SELECT DISTINCT(rating) 
FROM film
```

***(rating::varchar LIKE 'NC%' OR rating::varchar LIKE 'PG')***                                                                     
is treated like one seperate statement coz it's in parenthrsis.

```python
%%sql

SELECT *
FROM film
WHERE replacement_cost >= 10 AND replacement_cost <= 14
  AND  (rating::varchar LIKE 'NC%' OR rating::varchar LIKE 'PG')
LIMIT 10
```

**NOT** can be used like ^
- **NOT** BETWEEN
- **NOT** > or **NOT** <
- **NOT** LIKE
- **NOT** NULL
